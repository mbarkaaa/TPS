### **Question 1**

1- Importer les librairies : numpy, matplotlib, sklearn.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

### **Question 2**

2- Créer votre dataset en important la fonction datasets.make_regression et utilisez la pour générer un problème de régression aléatoire de 100 exemples avec une seule variable avec y=x^2

```python
from sklearn.datasets import make_regression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Générer le dataset de régression
X, y = make_regression(n_samples=100, n_features=1)

# Transformer les fonctionnalités en fonctionnalités polynomiales
poly = PolynomialFeatures(degree=2, include_bias=False)
X = poly.fit_transform(X)

# Remplacer y par y=x^2
y = y**2    #premiere colonne de la matrice X


```

### **Question 3**

3- Visualiser vos données en utilisant la fonction plt.scatter(x, y).

```python
import matplotlib.pyplot as plt
# Affichage du dataset généré
plt.scatter(X[:,0], y, color='blue')
plt.title('Dataset de régression (y=y^2)')
plt.xlabel('Variable indépendante (X)')
plt.ylabel('Variable dépendante (Y)')
plt.show()
```

### **Question 4**

4- Définir votre modèle avec SGDRegressor() sur 100 itérations avec un Learning rate de 0.0001.

```python
from sklearn.linear_model import SGDRegressor

# Définition du modèle avec SGDRegressor
modele_regression = SGDRegressor(max_iter=100, eta0=0.0001)
```

### **Question 5**

5- Entraîner votre modèle en utilisant la méthode fit().

```python
# Ajustement du modèle aux données
modele_regression.fit(X, y)
```

### **Question 6**

6- Calculer la précision de votre modèle en utilisant la fonction score qui calcule le coefficient de détermination entre le modèle et les valeurs y de votre Dataset.

```python
# Calcul de la précision du modèle
precision = modele_regression.score(X, y)

# Affichage de la précision
print("Précision du modèle (R^2) :", precision)
```

### **Question 7**

7- Faire de nouvelles prédictions avec la fonction predict() et tracer les résultats.

```python
# Faire de nouvelles prédictions
predictions = modele_regression.predict(X)

# Tracer les résultats
plt.scatter(X[:,0], y, color='blue', label='Données réelles')
plt.scatter(X[:,0], predictions, color='red', label='Prédictions du modèle', marker='+')
plt.title('Prédictions du modèle de régression')
plt.xlabel('Variable indépendante (X)')
plt.ylabel('Variable dépendante (Y)')
plt.legend()
plt.show()
```

### **Question 8**

8- Refaire le même travail en entraînant votre modèle sur 1000 itérations avec un Learning rate de 0.001.

```python
from sklearn.linear_model import SGDRegressor

# Définition du modèle avec SGDRegressor
modele_regression = SGDRegressor(max_iter=1000, learning_rate='constant', eta0=0.001, random_state=42)

# Ajustement du modèle aux données
modele_regression.fit(X, y)

# Calcul de la précision du modèle
precision = modele_regression.score(X, y)
print("Précision du modèle (R^2) :", precision)

# Faire de nouvelles prédictions
predictions = modele_regression.predict(X)

# Tracer les résultats
plt.scatter(X[:,0], y, color='blue', label='Données réelles')
plt.scatter(X[:,0], predictions, color='red', label='Prédictions du modèle', marker='+')
plt.title('Prédictions du modèle de régression (1000 itérations)')
plt.xlabel('Variable indépendante (X)')
plt.ylabel('Variable dépendante (Y)')
plt.legend()
plt.show()
```
