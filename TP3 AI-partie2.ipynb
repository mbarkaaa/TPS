```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
X, y = datasets.make_regression(n_samples=100, n_features=1, noise=0.1)
plt.scatter(X, y)
```

```python
model = SGDRegressor(max_iter=100, learning_rate='constant', eta0=0.0001)
```

```python
model.fit(X, y)
```

```python
precision = model.score(X, y)
print("Précision du modèle :", precision)
```

```python
y_pred = model.predict(X)
plt.scatter(X, y, label='Données réelles')
plt.plot(X, y_pred, color='red', label='Prédictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Prédicntions du modèle')
plt.legend()
plt.show()
```
